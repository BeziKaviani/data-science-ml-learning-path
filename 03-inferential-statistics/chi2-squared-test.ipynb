```python
import pandas as pd
import numpy as np
import scipy
import scipy.stats as sp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
employee_df = pd.read_csv("datasets/Employee.csv")
```

```python
employee_df
```

```python
ct = pd.crosstab(employee_df['Attrition'], employee_df['JobSatisfaction'], margins=True)
ct
```

```python
obs = np.append(ct.iloc[0, :4].values, ct.iloc[1, :-1].values)
obs
```

```python
row_sum = ct.iloc[:-1,-1].values
row_sum
```

```python
exp = []
for j in range(2):
    for value in ct.iloc[-1,:-1].values:
        exp.append(value * row_sum[j] / ct.loc["All", "All"])
```

```python
exp
```

```python
chi2_value = ((obs - exp) ** 2 / exp).sum()
```

Degree of freedome = (number of rows-1) * (number of columns-1)

```python
dof = ((len(row_sum) - 1) * (len(ct.iloc[2, :-1]) - 1))
dof
```

```python
sp.chi2.cdf(chi2_value, dof)
```

```python
p_value = 1 - sp.chi2.cdf(chi2_value, dof)
p_value
```
