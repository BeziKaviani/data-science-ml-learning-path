```python
import pandas as pd
import numpy as np
import scipy
import scipy.stats as sp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
employee_df = pd.read_csv("datasets/Employee.csv")
```

```python
ct = pd.crosstab(employee_df['Attrition'], employee_df['JobSatisfaction'], margins=True)
ct
```

```python
d = np.array([ct.iloc[0, :-1], ct.iloc[1, :-1]])
d
```

```python
sp.chi2_contingency(d)
```
