```python
import pandas as pd
import numpy as np
import scipy
import scipy.stats as sp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
employee_df = pd.read_csv("datasets/Employee.csv")
```

```python
ct_1 = pd.crosstab(employee_df['Attrition'], employee_df['JobSatisfaction'], margins=True)
ct_1
```

```python
d_1 = np.array([ct_1.iloc[0, :-1], ct_1.iloc[1, :-1]])
d_1
```

```python
sp.chi2_contingency(d_1)
```

```python
ct_2 = pd.crosstab(employee_df['Attrition'], employee_df["Education"], margins=True)
ct_2
```

```python
d = np.array([ct_2.iloc[0, :-1], ct_2.iloc[1, :-1]])
d
```
