```python
import pandas as pd
import numpy as np
import scipy
import scipy.stats as sp
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

Performing z formula on imaginary data

```python
sample_mean = 102
pop_mean = 100
std = 16
pop_count = 100
z = (sample_mean - pop_mean) / (std / math.sqrt(pop_count))
```

```python
z
```

```python

plt.fill_between(x = np.arange(-4, z, 0.01), y1 = sp.norm.pdf(np.arange(-4, z, 0.01)))
plt.fill_between(x = np.arange(z, 4, 0.01), y1 = sp.norm.pdf(np.arange(z, 4, 0.01)))
```
