```python
fig , ax = plt.subplots()
ax.boxplot(df["comb08"], labels=["All"])
ax.set(title="Vehicle MPG", ylabel="MPG for vehicle")
```

```python
fig , ax = plt.subplots()
ax.boxplot(df["comb08"], labels=["All"], showfliers=False)
ax.set(title="Vehicle MPG", ylabel="MPG for vehicle")
```

```python
fig , (ax1, ax2) = plt.subplots(2)
ax1.hist(df["comb08"], range=(10,100))
ax1.set(title="Vehicle MPG", xlabel="MPG for vehicle", ylabel="Number of vehicle")

ax2.boxplot(df["comb08"], labels=["All"], showfliers=False)
ax1.set(title="Vehicle MPG", ylabel="MPG for vehicle")
```

```python
fig , (ax1, ax2) = plt.subplots(2)
ax1.hist(df["comb08"], range=(10,100))
ax1.set(title="Vehicle MPG", xlabel="MPG for vehicle", ylabel="Number of vehicle")

ax2.boxplot(df["comb08"], labels=["All"], showfliers=False, vert=False)
ax2.set(ylabel="MPG for vehicle")

fig.suptitle("MPG Distribution", fontsize=14, fontweight="bold")
```

```python
fig , (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9,4))
ax1.hist(df["comb08"], range=(10,100))
ax1.set(title="Vehicle MPG",ylabel="Number of vehicle")

ax2.boxplot(df["comb08"], labels=["All"], showfliers=False, vert=False)
ax2.set(ylabel="MPG for vehicle")

fig.suptitle("MPG Distribution", fontsize=14, fontweight="bold")
```
